```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df = df.drop('customerID', axis=1)
df.head()
```

DATA PREPROCESSING:

Data preprocessing is a crucial step in model building, involving tasks like handling missing values, scaling features, and encoding categorical variables. This ensures the data is clean, standardized, and ready for effective modeling, ultimately enhancing the performance and accuracy of machine learning algorithms.

```python
from sklearn.preprocessing import MinMaxScaler
categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
df_encoded = pd.get_dummies(df, columns=categorical_columns)
df_encoded.head()
```

```python
df_encoded['Churn'] = df['Churn']
df_encoded['Churn'] = df_encoded['Churn'].map({'Yes': 1, 'No': 0})
df_encoded = df_encoded.drop(['Churn_No', 'Churn_Yes'], axis=1)
df_encoded = df_encoded.astype(int)
```

```python
df_encoded.head()
```

```python
x = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state= 98)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(x_train, y_train)
y_pred = logistic_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, precision_recall_fscore_support, precision_recall_curve
accuracy = accuracy_score(y_test, y_pred)
classification_report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy}\nClassification Report:\n{classification_report}")
```

INTERPRETATION:

The model exhibits an overall accuracy of 81.07%, demonstrating its effectiveness in correctly predicting customer churn. While precision for the positive class (Churn = 1) is 63%, indicating moderate reliability, the recall of 55% suggests room for improvement in capturing all actual instances of churn.
